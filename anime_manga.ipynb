Import libs

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import Dataset

```python
manga_data=pd.read_csv('F:\\Vscode\\Anime_and_Manga\\myanimelist-anime-and-manga\\MAL-manga.csv')
anime_data=pd.read_csv('F:\\Vscode\Anime_and_Manga\\myanimelist-anime-and-manga\\MAL-anime.csv')
```

Columns changes

```python
manga_data.columns=manga_data.columns.str.replace(" ","_")
anime_data.columns=anime_data.columns.str.replace(" ","_")
```

```python
manga_data.head()
```

Repalcing few Data

```python
manga_data['Volumes'] = manga_data['Volumes'].replace(['?'], 'Not Known')
anime_data['Episodes']=anime_data['Episodes'].replace(['?'], 'Not Known')
```

Using labda Function 

```python
anime_data['Status'] = anime_data['Aired'].apply(lambda x: "Finished" if x.split('-')[1] else "Ongoing")
```

Collecting Valuable data only

```python
anime_data = anime_data[['Unnamed:_0','Title','Rank','Type','Episodes','Aired','Members','Score']]
manga_data = manga_data[['Unnamed:_0','Title','Rank','Type','Volumes','Published','Members','Score']]
```

Renaming column

```python
anime_data.rename(columns = {'Unnamed:_0':'id'}, inplace = True)
manga_data.rename(columns = {'Unnamed:_0':'id'}, inplace = True)
```

correlation

```python
manga_data_correlation=manga_data.corr(numeric_only=True)
```

Manga -- Data -- Type ---> Score

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Type', y='Score', data=manga_data)
plt.title("TYPE -- SCORE")
plt.xlabel("Type")
plt.ylabel("Score")
plt.show()
```

Volume -- Score

```python
plt.figure(figsize=(12, 12))
sns.barplot(y='Volumes', x='Score', data=manga_data)
plt.title("Volumes -- SCORE")
plt.xlabel("Volumes")
plt.ylabel("Score")
plt.show()
```

TOP MANGA

```python
most_score_ = manga_data.groupby('Title')['Score'].sum().nlargest(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=most_score_.index, y=most_score_.values)
plt.xticks(rotation=90)
plt.title("Top Manga")
plt.xlabel("Title")
plt.ylabel("Score")
plt.show()
```

Published -- Score

```python
most_Published_ = manga_data.groupby('Published')['Score'].sum().nlargest(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=most_Published_.index, y=most_Published_.values)
plt.xticks(rotation=90)
plt.title("TOP - Published")
plt.xlabel("Published")
plt.ylabel("Score")
plt.show()
```

Top Members score

```python
most_Members_ = manga_data.groupby('Members')['Score'].sum().nlargest(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=most_Members_.index, y=most_Members_.values)
plt.xticks(rotation=90)
plt.title("TOP - Members")
plt.xlabel("Members")
plt.ylabel("Score")
plt.show()
```
